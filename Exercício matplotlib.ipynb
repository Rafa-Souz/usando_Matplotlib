6)** Agora use x, y e z arrays para criar um gráfico. Configure os limites xlimits e y no gráfico inserido: **

```python
fig = plt.figure()
axx1 = fig.add_axes([0,0,1,1])
axx2 = fig.add_axes([0.2,0.5,0.4,0.4])

axx1.plot(x,z)
axx2.plot(x,y)

axx2.set_xlim(20 ,22)
axx2.set_ylim(30, 50)
#o código xlim e ylim é para especificar a leitura de uma certo trecho do gráfico.

axx1.set_xlabel('X')
axx1.set_ylabel('Z')

axx2.set_xlabel('X')
axx2.set_ylabel('Y')
axx2.set_title('Zoom')
```

8)** Agora trace (x, y) e (x, z) nos eixos. Note a largura de linha e o estilo objetivos. **

```python
plt.subplots(1,2)
plt.subplot(1,2,1)
plt.plot(x,y, 'b--', linewidth=4, alpha= 1)

plt.subplot(1,2,2)
plt.plot(x,z, 'r-', linewidth=4, alpha= 1)
```

9)** Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () apenas copiando e colando seu código anterior. **

```python
fig, axes= plt.subplots(nrows= 1, ncols=2, figsize=(12,2))

axes [0].plot(x,y, 'b-', linewidth=5)
axes [1].plot(x,z, 'r--', linewidth=2)
```
